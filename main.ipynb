<b>5. Bagaimana strategi retensi dapat dirancang?<b>

```python
desc_stats = df_copy_cleaned[['MonthlyCharges', 'tenure']].describe()
print("\nStatistik Deskriptif:\n", desc_stats)
```

```python
var_col = ['MonthlyCharges', 'tenure']
for col in df_copy_cleaned[var_col]:
    variance = df_copy_cleaned[col].var()
    print(f'Variance of {col} data:', variance)
```

```python
Q1 = df_copy_cleaned['MonthlyCharges'].quantile(0.25)
Q3 = df_copy_cleaned['MonthlyCharges'].quantile(0.75)
IQR = Q3 - Q1
outliers = df_copy_cleaned[(df_copy_cleaned['MonthlyCharges'] < (Q1 - 1.5 * IQR)) | (df_copy_cleaned['MonthlyCharges'] > (Q3 + 1.5 * IQR))]
print(f"\nJumlah Outlier MonthlyCharges: {len(outliers)}")
```

Berdasarkan analisa statistik deskriptif diatas, strategi retensi bisa diarahkan pada dua kelompok besar pelanggan: mereka yang membayar tinggi dan mereka yang berada di awal masa berlangganan.

Rata-rata biaya bulanan sekitar $64,8 dengan variasi yang cukup besar (variance ±905), menunjukkan adanya rentang harga yang luas. Pelanggan dengan Monthly Charges tinggi, khususnya yang berada di kuartil atas (di atas ±$90), menjadi kelompok yang perlu perhatian khusus karena dari analisis sebelumnya, tarif tinggi berkorelasi dengan tingkat churn yang lebih besar. Strategi untuk mereka bisa berupa program diskon loyalitas, paket bundling yang menambah nilai tanpa menaikkan harga, atau personalisasi layanan agar mereka merasa biaya yang dibayar sepadan.

Sementara itu, rata-rata tenure hanya sekitar 32 bulan dengan distribusi yang lebar (variance ±602), dan kelompok dengan tenure rendah (kurang dari 12 bulan) adalah yang paling rawan churn. Untuk pelanggan baru, retensi dapat difokuskan pada masa-masa awal dengan memberikan pengalaman onboarding yang kuat, penawaran khusus untuk tahun pertama, dan komunikasi yang lebih intensif untuk membangun keterikatan.

Karena data tidak menunjukkan outlier pada biaya bulanan, ini menandakan tarif yang ada relatif konsisten dan penyesuaian harga untuk strategi retensi bisa dilakukan secara terukur tanpa mengganggu struktur harga keseluruhan.

Strategi ini dimaksudkan untuk merangkul pelanggan yang baru supaya betah, dan meyakinkan pelanggan yang lama bahwa biaya yang mereka bayar layak untuk diteruskan sehingga menciptakan keberlanjutan.

<b>6. Apakah jenis kontrak memengaruhi churn secara signifikan?<b>

```python
contingency_table = pd.crosstab(df_copy_cleaned['Contract'], df_copy_cleaned['Churn'])
contingency_table
```

H0: Tidak terdapat hubungan yang signifikan antara jenis kontrak pelanggan dan status churn (Proporsi churn pada pelanggan Month-to-month, One year, dan Two year adalah sama).

H1: Terdapat hubungan yang signifikan antara jenis kontrak pelanggan dan status churn (proporsi churn berbeda tergantung jenis kontraknya).

```python
res = stats.chi2_contingency(contingency_table)
print("P-value:",res.pvalue)

if res.pvalue < 0.05:
    print("-> Hasil signifikan: Ada perbedaan proporsi churn antar tipe kontrak.")
else:
    print("-> Tidak signifikan: Tidak ada bukti perbedaan proporsi churn antar tipe kontrak.")
```

Berdasarkan hasil analisis statsitik inferensial mengenai keterkaitan jenis kontrak dengan churn melalui uji chi-square menunjukkan p-value yang sangat kecil, jauh di bawah ambang (treshold) 0,05. Artinya, kita dapat menolak hipotesis nol dan menyimpulkan bahwa jenis kontrak memang berpengaruh signifikan terhadap churn.

Berdasarkan informasi dari tabel kontingensi, pelanggan dengan kontrak bulan-ke-bulan memiliki proporsi churn yang jauh lebih tinggi (1655 churn dari total 3875 pelanggan, atau sekitar 42,7%) dibandingkan kontrak satu tahun (166 churn dari 1472 pelanggan, sekitar 11,3%) maupun kontrak dua tahun (48 churn dari 1685 pelanggan, sekitar 2,8%).
